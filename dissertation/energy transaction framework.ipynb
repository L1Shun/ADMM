# 3. Without ADMM #

```python
"""
Define distribution network (DN) part
====================================================================================================================================================================================
"""
# Define the total DN cost 
C_DN_WS_Tran = Lamb_DN_WS_buy * E_DN_WS_buy - Lamb_DN_WS_sell * E_DN_WS_sell                   # Transaction cost between DN and WS
C_DN_MG_Tran = Lamb_DN_MG_buy * E_DN_MG1_buy - Lamb_DN_MG_sell * E_DN_MG1_sell + Lamb_DN_MG_buy * E_DN_MG2_buy \
- Lamb_DN_MG_sell * E_DN_MG2_sell + Lamb_DN_MG_buy * E_DN_MG3_buy - Lamb_DN_MG_sell * E_DN_MG3_sell # Transaction cost between DN and MG,m
C_DN_e_Tran =  C_DN_WS_Tran + C_DN_MG_Tran                                                     # Total electricity transaction cost 

C_DG_DN = a * E_DG_DN ** 2 + b * E_DG_DN + c                                                   # Gas cost by DG
C_MT_DN = (Cnl/L) * (E_MT_DN/eff_MT)                                                           # Gas cost by MT
C_FC_DN = (Cnl/L) * (E_FC_DN/eff_FC)                                                           # Gas cost by FC
C_GB1_DN = (Cnl/L) * (H_GB1_DN/eff_GB)                                                         # Gas cost by GB
C_GB2_DN = (Cnl/L) * (H_GB2_DN/eff_GB)                                                         # Gas cost by GB
C_CHP1_DN = (Cnl/L) * ((E_CHP1_DN/eff_e + H_CHP1_DN/(eff_rec * (1-eff_e))))/eff_b              # Gas cost by CHP
C_CHP2_DN = (Cnl/L) * ((E_CHP2_DN/eff_e + H_CHP2_DN/(eff_rec * (1-eff_e))))/eff_b              # Gas cost by CHP
C_DN_h_g = C_DG_DN + C_MT_DN + C_FC_DN + C_GB1_DN + C_GB2_DN + C_CHP1_DN + C_CHP2_DN           # Total Cost of gas use to generate electricity in DN

C_DN_h_Tran = Lamb_DN_GN_buy * H_DN_GN_buy - Lamb_DN_GN_sell * H_DN_GN_sell                    # Heat transaction cost

C_DN_e = C_DN_e_Tran                                                                           # Electricity cost for distribution grid
C_DN_h = C_DN_h_g + C_DN_h_Tran                                                                # Heating cost for distribution grid

DNcost = C_DN_e + C_DN_h                                                                       # Total cost
'''
# Define First Microgrid (MG1) part, while PV, WT and CHP are suppliers of electrical power for MG1
====================================================================================================================================================================================
====================================================================================================================================================================================
'''
# Define the total MG1 cost 
C_MG1_e_Tran = Lamb_MG_DN_buy * E_MG1_DN_buy - Lamb_MG_DN_sell * E_MG1_DN_sell               # Transaction cost between MG1 and DN 

C_MT_MG1 = (Cnl/L) * (E_MT_MG1 / eff_MT)                                                
C_GB1_MG1 = (Cnl/L) * (H_GB1_MG1/eff_GB)                                                     # Total Cost of gas use to generate electricity in MG1
C_GB2_MG1 = (Cnl/L) * (H_GB2_MG1/eff_GB)
C_CHP_MG1 = (Cnl/L) * ((E_CHP_MG1/eff_e + H_CHP_MG1/(eff_rec * (1-eff_e))))/eff_b            # Gas cost by CHP
C_MG1_h_g = C_GB1_MG1 + C_GB2_MG1 + C_CHP_MG1 + C_MT_MG1

C_MG1_h_Tran = Lamb_MG_GN_buy * H_MG1_GN_buy - Lamb_MG_GN_sell * H_MG1_GN_sell               # Heat transaction cost

C_MG1_e = C_MG1_e_Tran                                                                       # Electricity cost for MG1
C_MG1_h = C_MG1_h_g + C_MG1_h_Tran                                                           # Heating cost for MG1

MG1cost = C_MG1_e + C_MG1_h                                                                  # Total cost
'''
# Define First Microgrid (MG2) part, while MT, WT and CHP are suppliers of electrical power for MG2
====================================================================================================================================================================================
====================================================================================================================================================================================
'''

# Define the objective
C_MG2_e_Tran = Lamb_MG_DN_buy * E_MG2_DN_buy - Lamb_MG_DN_sell * E_MG2_DN_sell               # Transaction cost between MG2 and DN 
                                            
C_GB1_MG2 = (Cnl/L) * (H_GB1_MG2/eff_GB)                                                     # Total Cost of gas use to generate electricity in MG2
C_GB2_MG2 = (Cnl/L) * (H_GB2_MG2/eff_GB)
C_MT_MG2 = (Cnl/L) * (E_MT_MG2/eff_MT)                                                 
C_CHP_MG2 = (Cnl/L) * ((E_CHP_MG2/eff_e + H_CHP_MG2/(eff_rec * (1-eff_e))))/eff_b            # Gas cost by CHP
C_MG2_h_g = C_GB1_MG2 + C_GB2_MG2 + C_CHP_MG2 + C_MT_MG2 

C_MG2_h_Tran = Lamb_MG_GN_buy * H_MG2_GN_buy - Lamb_MG_GN_sell * H_MG2_GN_sell               # Heat transaction cost


C_MG2_e = C_MG2_e_Tran                                                     # Electricity cost for MG2
C_MG2_h = C_MG2_h_g + C_MG2_h_Tran                                        # Heating cost for MG2

MG2cost = C_MG2_e + C_MG2_h                                                                  # Total cost
'''
# Define First Microgrid (MG3) part, while PV and CHP are suppliers of electrical power for MG3
====================================================================================================================================================================================
====================================================================================================================================================================================
'''
C_MG3_e_Tran = Lamb_MG_DN_buy * E_MG3_DN_buy - Lamb_MG_DN_sell * E_MG3_DN_sell               # Transaction cost between MG3 and DN 

C_FC_MG3 = (Cnl/L) * (E_FC_MG3/eff_FC)                                               
C_GB1_MG3 = (Cnl/L) * (H_GB1_MG3/eff_GB)                                                     # Total Cost of gas use to generate electricity in MG3
C_GB2_MG3 = (Cnl/L) * (H_GB2_MG3/eff_GB)                                                        
C_CHP_MG3 = (Cnl/L) * ((E_CHP_MG3/eff_e + H_CHP_MG3/(eff_rec * (1-eff_e))))/eff_b            # Gas cost by CHP
C_MG3_h_g = C_GB1_MG3 + C_GB2_MG3 + C_CHP_MG3 + C_FC_MG3

C_MG3_h_Tran = Lamb_MG_GN_buy * H_MG3_GN_buy - Lamb_MG_GN_sell * H_MG3_GN_sell               # Heat transaction cost

C_MG3_e = C_MG3_e_Tran                                                                       # Electricity cost for MG3
C_MG3_h = C_MG3_h_g + C_MG3_h_Tran                                                           # Heating cost for MG3

MG3cost = C_MG3_e + C_MG3_h                                                                  # Total cost


'''
====================================================================================================================================================================================
'''

# Define the objective
cost = DNcost + MG1cost + MG2cost + MG3cost
objective = cp.Minimize(cp.sum(cost)) # Total cost with electricity and heating parts

# Define the constraints
constraints = [ E_DN_WS_buy >= MIN_E_DN_WS_Tran * I_E_DN_WS_Tran, E_DN_WS_buy <= MAX_E_DN_WS_Tran * I_E_DN_WS_Tran,                 # min and max of electricity transaction with WS   
                E_DN_WS_sell >= MIN_E_DN_WS_Tran * (1-I_E_DN_WS_Tran), E_DN_WS_sell <= MAX_E_DN_WS_Tran * (1-I_E_DN_WS_Tran),

                E_DN_MG1_buy >= MIN_E_DN_MG1_Tran * I_E_DN_MG1_Tran, E_DN_MG1_buy <= MAX_E_DN_MG1_Tran * I_E_DN_MG1_Tran,           # min and max of electricity transaction with MG1           
                E_DN_MG1_sell >= MIN_E_DN_MG1_Tran * (1-I_E_DN_MG1_Tran), E_DN_MG1_sell <= MAX_E_DN_MG1_Tran * (1-I_E_DN_MG1_Tran),

                E_DN_MG2_buy >= MIN_E_DN_MG2_Tran * I_E_DN_MG2_Tran, E_DN_MG2_buy <= MAX_E_DN_MG2_Tran * I_E_DN_MG2_Tran,           # min and max of electricity transaction with MG2          
                E_DN_MG2_sell >= MIN_E_DN_MG2_Tran * (1-I_E_DN_MG2_Tran), E_DN_MG2_sell <= MAX_E_DN_MG2_Tran * (1-I_E_DN_MG2_Tran),

                E_DN_MG3_buy >= MIN_E_DN_MG3_Tran * I_E_DN_MG3_Tran, E_DN_MG3_buy <= MAX_E_DN_MG3_Tran * I_E_DN_MG3_Tran,           # min and max of electricity transaction with MG3           
                E_DN_MG3_sell >= MIN_E_DN_MG3_Tran * (1-I_E_DN_MG3_Tran), E_DN_MG3_sell <= MAX_E_DN_MG3_Tran * (1-I_E_DN_MG3_Tran),

                E_EES_ch_DN >= MIN_EES_ch * I_DN_EES, E_EES_ch_DN <= MAX_EES_ch * I_DN_EES,                                         # EES
                E_EES_dch_DN >= MIN_EES_dch * (1 - I_DN_EES), E_EES_dch_DN <= MAX_EES_dch * (1 - I_DN_EES),
                E_EES_DN[0] == 0,                                                                                                   # Set the energy level of EES
                E_EES_DN[1:] == E_EES_DN[:-1] + eff_EES_ch * E_EES_ch_DN - E_EES_dch_DN / eff_EES_dch,
                E_EES_DN >= MIN_E_EES, E_EES_DN <= MAX_E_EES,

                E_DG_DN >= MIN_E_DG_DN * I_DN_DG, E_DG_DN <= MAX_E_DG_DN * I_DN_DG,                                                 # min and max of electricity generated by DG                   
                                                                   
                E_MT_DN >= MIN_E_MT_DN * I_DN_MT, E_MT_DN <= MAX_E_MT_DN * I_DN_MT,                                                 # min and max of electricity generated by MT
                                                                  
                E_FC_DN >= MIN_E_FC_DN * I_DN_FC, E_FC_DN <= MAX_E_FC_DN * I_DN_FC,                                                 # min and max of electricity generated by FC
                                                           
                E_CHP1_DN >= MIN_E_CHP1_DN * I_DN_CHP1, E_CHP1_DN <= MAX_E_CHP1_DN * I_DN_CHP1,                                     # min and max of electricity generated by CHP1                                  
                H_CHP1_DN >= MIN_H_CHP1_DN * I_DN_CHP1, H_CHP1_DN <= MAX_H_CHP1_DN * I_DN_CHP1,                                     # min and max of heating generated by CHP1
                E_CHP1_DN / V == H_CHP1_DN / (1 - V),                                                                               # Proportion between electricity and heat generated by CHP1
                                                            
                E_CHP2_DN >= MIN_E_CHP2_DN * I_DN_CHP2, E_CHP2_DN <= MAX_E_CHP2_DN * I_DN_CHP2,                                     # min and max of electricity generated by CHP2                                                          
                H_CHP2_DN >= MIN_H_CHP2_DN * I_DN_CHP2, H_CHP2_DN <= MAX_H_CHP2_DN * I_DN_CHP2,                                     # min and max of heating generated by CHP2
                E_CHP2_DN / V == H_CHP2_DN / (1 - V),                                                                               # Proportion between electricity and heat generated by CHP2
                                                               
                H_GB1_DN >= MIN_H_GB_DN * I_DN_GB1, H_GB1_DN <= MAX_H_GB_DN * I_DN_GB1,                                             # min and max of heating generated by GB1
                H_GB2_DN >= MIN_H_GB_DN * I_DN_GB2, H_GB2_DN <= MAX_H_GB_DN * I_DN_GB2,                                             # min and max of heating generated by GB2
 
                H_DN_GN_buy >= MIN_H_DN_GN_Tran * I_H_DN_GN_Tran, H_DN_GN_buy <= MAX_H_DN_GN_Tran * I_H_DN_GN_Tran,                 # min and max of heating transaction  
                H_DN_GN_sell >= MIN_H_DN_GN_Tran * (1-I_H_DN_GN_Tran), H_DN_GN_sell <= MAX_H_DN_GN_Tran * (1-I_H_DN_GN_Tran),

                H_HS_ch_DN >= MIN_HS_ch * I_DN_HS, H_HS_ch_DN <= MAX_HS_ch * I_DN_HS,                                               # HS
                H_HS_dch_DN >= MIN_HS_dch * (1 - I_DN_HS), H_HS_dch_DN <= MAX_HS_dch * (1 - I_DN_HS),
                H_HS_DN[0] == 0,                                                                                                    # Set the energy level of HS
                H_HS_DN[1:] == H_HS_DN[:-1] + eff_HS_ch * H_HS_ch_DN - H_HS_dch_DN / eff_HS_dch,
                H_HS_DN >= MIN_H_HS, H_HS_DN <= MAX_H_HS,

                O_EC_DN == E_EC_DN / eff_EC,
                O_EC_DN >= MIN_O_EC_DN * I_DN_EC,
                O_EC_DN <= MAX_O_EC_DN * I_DN_EC,

                O_AC_DN == H_AC_DN / eff_AC,
                O_AC_DN >= MIN_O_AC_DN * I_DN_AC,
                O_AC_DN <= MAX_O_AC_DN * I_DN_AC,  

                (E_DN_WS_buy - E_DN_WS_sell) + (E_DN_MG1_buy - E_DN_MG1_sell) + (E_DN_MG2_buy - E_DN_MG2_sell) + (E_DN_MG3_buy - E_DN_MG3_sell) 
                +(E_EES_dch_DN - E_EES_ch_DN) + E_MT_DN + E_FC_DN + E_CHP1_DN + E_CHP2_DN + E_PV_DN + E_WT_DN - E_EC_DN == EL_DN,             # Electricity load of DN

                (H_DN_GN_buy - H_DN_GN_sell) + (H_HS_dch_DN - H_HS_ch_DN) + H_CHP1_DN + H_CHP2_DN + H_GB1_DN + H_GB2_DN - H_AC_DN == HL_DN,   # Heating load of DN

                O_EC_DN + O_AC_DN == OL_DN,

                E_MG1_DN_buy >= MIN_E_MG_DN_Tran * I_E_MG1_DN_Tran, E_MG1_DN_buy <= MAX_E_MG_DN_Tran * I_E_MG1_DN_Tran,                 # min and max of electricity transaction with DN           
                E_MG1_DN_sell >= MIN_E_MG_DN_Tran * (1-I_E_MG1_DN_Tran), E_MG1_DN_sell <= MAX_E_MG_DN_Tran * (1-I_E_MG1_DN_Tran),

                E_EES_ch_MG1 >= MIN_EES_ch * I_MG1_EES, E_EES_ch_MG1 <= MAX_EES_ch * I_MG1_EES,                                         # EES
                E_EES_dch_MG1 >= MIN_EES_dch * (1 - I_MG1_EES), E_EES_dch_MG1 <= MAX_EES_dch * (1 - I_MG1_EES),
                E_EES_MG1 >= MIN_E_EES, E_EES_MG1 <= MAX_E_EES,
                E_EES_MG1[0] == 0,                                                                                                      # Set the energy level of EES
                E_EES_MG1[1:] == E_EES_MG1[:-1] + eff_EES_ch * E_EES_ch_MG1 - E_EES_dch_MG1 / eff_EES_dch,
                                                                                                      
                E_CHP_MG1 >= MIN_E_CHP_MG1 * I_MG1_CHP, E_CHP_MG1 <= MAX_E_CHP_MG1 * I_MG1_CHP,                                         # min and max of electricity generated by CHP                                                              
                H_CHP_MG1 >= MIN_H_CHP_MG1 * I_MG1_CHP, H_CHP_MG1 <= MAX_H_CHP_MG1 * I_MG1_CHP,                                         # min and max of heating generated by CHP
                E_CHP_MG1 / V == H_CHP_MG1 / (1 - V),                                                                                   # Proportion between electricity and heat generated by CHP

                E_MT_MG1 >= MIN_E_MT_MG1 * I_MG1_MT,
                E_MT_MG1 <= MAX_E_MT_MG1 * I_MG1_MT,  

                H_GB1_MG1 >= MIN_H_GB_MG * I_MG1_GB1, H_GB1_MG1 <= MAX_H_GB_MG * I_MG1_GB1,                                             # min and max of heating generated by GB1                    
                H_GB2_MG1 >= MIN_H_GB_MG * I_MG1_GB2, H_GB2_MG1 <= MAX_H_GB_MG * I_MG1_GB2,                                              # min and max of heating generated by GB2                 
                    
                H_MG1_GN_buy >= MIN_H_MG1_GN_Tran * I_H_MG1_GN_Tran, H_MG1_GN_buy <= MAX_H_MG1_GN_Tran * I_H_MG1_GN_Tran,               # min and max of heating transaction    
                H_MG1_GN_sell >= MIN_H_MG1_GN_Tran * (1-I_H_MG1_GN_Tran), H_MG1_GN_sell <= MAX_H_MG1_GN_Tran * (1-I_H_MG1_GN_Tran),

                H_HS_ch_MG1 >= MIN_HS_ch * I_MG1_HS, H_HS_ch_MG1 <= MAX_HS_ch * I_MG1_HS,                                               # HS
                H_HS_dch_MG1 >= MIN_HS_dch * (1 - I_MG1_HS), H_HS_ch_MG1 <= MAX_HS_dch * (1 - I_MG1_HS),
                H_HS_MG1[0] == 0,                                                                                                       # Set the energy level of HS
                H_HS_MG1[1:] == H_HS_MG1[:-1] + eff_HS_ch * H_HS_ch_MG1 - H_HS_dch_MG1 / eff_HS_dch,
                H_HS_MG1 >= MIN_H_HS, H_HS_MG1 <= MAX_H_HS,

                O_EC_MG1 == E_EC_MG1 / eff_EC,
                O_EC_MG1 >= MIN_O_EC_MG1 * I_MG1_EC,
                O_EC_MG1 <= MAX_O_EC_MG1 * I_MG1_EC,

                O_AC_MG1 == H_AC_MG1 / eff_AC,
                O_AC_MG1 >= MIN_O_AC_MG1 * I_MG1_AC,
                O_AC_MG1 <= MAX_O_AC_MG1 * I_MG1_AC,  

                (E_MG1_DN_buy - E_MG1_DN_sell) + (E_EES_dch_MG1 - E_EES_ch_MG1) + E_CHP_MG1 + E_PV_MG1 + E_WT_MG1 - E_EC_MG1 == EL_MG1,            # Electricity load of DN
                (H_MG1_GN_buy - H_MG1_GN_sell) + (H_HS_dch_MG1 - H_HS_ch_MG1) + H_CHP_MG1 + H_GB1_MG1 + H_GB2_MG1 - H_AC_MG1 == HL_MG1,             # Heating load of DN
                O_EC_MG1 + O_AC_MG1 == OL_MG1,

                E_MG2_DN_buy >= MIN_E_MG_DN_Tran * I_E_MG2_DN_Tran, E_MG2_DN_buy <= MAX_E_MG_DN_Tran * I_E_MG2_DN_Tran,                 # min and max of electricity transaction with DN           
                E_MG2_DN_sell >= MIN_E_MG_DN_Tran * (1-I_E_MG2_DN_Tran), E_MG2_DN_sell <= MAX_E_MG_DN_Tran * (1-I_E_MG2_DN_Tran),

                E_EES_ch_MG2 >= MIN_EES_ch * I_MG2_EES, E_EES_ch_MG2 <= MAX_EES_ch * I_MG2_EES,                                         # EES
                E_EES_dch_MG2 >= MIN_EES_dch * (1 - I_MG2_EES), E_EES_dch_MG2 <= MAX_EES_dch * (1 - I_MG2_EES),
                E_EES_MG2[0] == 0,                                                                                                      # Set the energy level of EES
                E_EES_MG2[1:] == E_EES_MG2[:-1] + eff_EES_ch * E_EES_ch_MG2 - E_EES_dch_MG2 / eff_EES_dch,
                E_EES_MG2 >= MIN_E_EES, E_EES_MG2 <= MAX_E_EES,

                E_MT_MG2 >= MIN_E_MT_MG2 * I_MG2_MT, E_MT_MG2 <= MAX_E_MT_MG2 * I_MG2_MT,                                               # min and max of electricity generated by MT     

                E_CHP_MG2 >= MIN_E_CHP_MG2 * I_MG2_CHP, E_CHP_MG2 <= MAX_E_CHP_MG2 * I_MG2_CHP,                                         # min and max of electricity generated by CHP                                                              
                H_CHP_MG2 >= MIN_H_CHP_MG2 * I_MG2_CHP, H_CHP_MG2 <= MAX_H_CHP_MG2 * I_MG2_CHP,                                         # min and max of heating generated by CHP
                E_CHP_MG2 / V == H_CHP_MG2 / (1 - V),                                                                                   # Proportion between electricity and heat generated by CHP
                                                                 
                H_GB1_MG2 >= MIN_H_GB_MG * I_MG2_GB1, H_GB1_MG2 <= MAX_H_GB_MG * I_MG2_GB1,                                             # min and max of heating generated by GB1
                H_GB2_MG2 >= MIN_H_GB_MG * I_MG2_GB2, H_GB2_MG2 <= MAX_H_GB_MG * I_MG2_GB2,                                             # min and max of heating generated by GB2  
                 
                H_MG2_GN_buy >= MIN_H_MG2_GN_Tran * I_H_MG2_GN_Tran, H_MG2_GN_buy <= MAX_H_MG2_GN_Tran * I_H_MG2_GN_Tran,               # min and max of heating transaction    
                H_MG2_GN_sell >= MIN_H_MG2_GN_Tran * (1-I_H_MG2_GN_Tran), H_MG2_GN_sell <= MAX_H_MG2_GN_Tran * (1-I_H_MG2_GN_Tran),

                H_HS_ch_MG2 >= MIN_HS_ch * I_MG2_HS, H_HS_ch_MG2 <= MAX_HS_ch * I_MG2_HS,                                               # HS
                H_HS_dch_MG2 >= MIN_HS_dch * (1 - I_MG2_HS), H_HS_ch_MG2 <= MAX_HS_dch * (1 - I_MG2_HS),
                H_HS_MG2[0] == 0,                                                                                                       # Set the energy level of HS
                H_HS_MG2[1:] == H_HS_MG2[:-1] + eff_HS_ch * H_HS_ch_MG2 - H_HS_dch_MG2 / eff_HS_dch,
                H_HS_MG2 >= MIN_H_HS, H_HS_MG2 <= MAX_H_HS,

                O_EC_MG2 == E_EC_MG2 / eff_EC,
                O_EC_MG2 >= MIN_O_EC_MG2 * I_MG2_EC,
                O_EC_MG2 <= MAX_O_EC_MG2 * I_MG2_EC,

                O_AC_MG2 == H_AC_MG2 / eff_AC,
                O_AC_MG2 >= MIN_O_AC_MG2 * I_MG2_AC,
                O_AC_MG2 <= MAX_O_AC_MG2 * I_MG2_AC, 

                (E_MG2_DN_buy - E_MG2_DN_sell) + (E_EES_dch_MG2 - E_EES_ch_MG2) + E_CHP_MG2 + E_MT_MG2 + E_WT_MG2 - E_EC_MG2 == EL_MG2,            # Electricity load of MG2
                (H_MG2_GN_buy - H_MG2_GN_sell) + (H_HS_dch_MG2 - H_HS_ch_MG2) + H_CHP_MG2 + H_GB1_MG2 + H_GB2_MG2 - H_AC_MG2 == HL_MG2,            # Heating load of MG2
                O_EC_MG2 + O_AC_MG2 == OL_MG2,

                E_MG3_DN_buy >= MIN_E_MG_DN_Tran * I_E_MG3_DN_Tran, E_MG3_DN_buy <= MAX_E_MG_DN_Tran * I_E_MG3_DN_Tran,                 # min and max of electricity transaction with DN           
                E_MG3_DN_sell >= MIN_E_MG_DN_Tran * (1-I_E_MG3_DN_Tran), E_MG3_DN_sell <= MAX_E_MG_DN_Tran * (1-I_E_MG3_DN_Tran),

                E_EES_ch_MG3 >= MIN_EES_ch * I_MG3_EES, E_EES_ch_MG3 <= MAX_EES_ch * I_MG3_EES,                                         # EES
                E_EES_dch_MG3 >= MIN_EES_dch * (1 - I_MG3_EES), E_EES_dch_MG3 <= MAX_EES_dch * (1 - I_MG3_EES),
                E_EES_MG3[0] == 0,                                                                                                      # Set the energy level of EES
                E_EES_MG3[1:] == E_EES_MG3[:-1] + eff_EES_ch * E_EES_ch_MG3 - E_EES_dch_MG3 / eff_EES_dch,
                E_EES_MG3 >= MIN_E_EES, E_EES_MG3 <= MAX_E_EES,  
                
                E_FC_MG3 >= MIN_E_FC_MG3 * I_MG3_FC,
                E_FC_MG3 <= MAX_E_FC_MG3 * I_MG3_FC,  

                E_CHP_MG3 >= MIN_E_CHP_MG3 * I_MG3_CHP, E_CHP_MG3 <= MAX_E_CHP_MG3 * I_MG3_CHP,                                         # min and max of electricity generated by CHP                                                              
                H_CHP_MG3 >= MIN_H_CHP_MG3 * I_MG3_CHP, H_CHP_MG3 <= MAX_H_CHP_MG3 * I_MG3_CHP,                                         # min and max of heating generated by CHP
                E_CHP_MG3 / V == H_CHP_MG3 / (1 - V),                                                                                   # Proportion between electricity and heat generated by CHP
                                                                 
                H_GB1_MG3 >= MIN_H_GB_MG * I_MG3_GB1, H_GB1_MG3 <= MAX_H_GB_MG * I_MG3_GB1,                                             # min and max of heating generated by GB1
                H_GB2_MG3 >= MIN_H_GB_MG * I_MG3_GB2, H_GB2_MG3 <= MAX_H_GB_MG * I_MG3_GB2,                                             # min and max of heating generated by GB2  
                 
                H_MG3_GN_buy >= MIN_H_MG3_GN_Tran * I_H_MG3_GN_Tran, H_MG3_GN_buy <= MAX_H_MG3_GN_Tran * I_H_MG3_GN_Tran,               # min and max of heating transaction    
                H_MG3_GN_sell >= MIN_H_MG3_GN_Tran * (1-I_H_MG3_GN_Tran), H_MG3_GN_sell <= MAX_H_MG3_GN_Tran * (1-I_H_MG3_GN_Tran),

                H_HS_ch_MG3 >= MIN_HS_ch * I_MG3_HS, H_HS_ch_MG3 <= MAX_HS_ch * I_MG3_HS,                                               # HS
                H_HS_dch_MG3 >= MIN_HS_dch * (1 - I_MG3_HS), H_HS_ch_MG3 <= MAX_HS_dch * (1 - I_MG3_HS),
                H_HS_MG3[0] == 0,                                                                                                       # Set the energy level of HS
                H_HS_MG3[1:] == H_HS_MG3[:-1] + eff_HS_ch * H_HS_ch_MG3 - H_HS_dch_MG3 / eff_HS_dch,
                H_HS_MG3 >= MIN_H_HS, H_HS_MG3 <= MAX_H_HS,

                O_EC_MG3 == E_EC_MG3 / eff_EC,
                O_EC_MG3 >= MIN_O_EC_MG3 * I_MG3_EC,
                O_EC_MG3 <= MAX_O_EC_MG3 * I_MG3_EC,

                O_AC_MG3 == H_AC_MG3 / eff_AC,
                O_AC_MG3 >= MIN_O_AC_MG3 * I_MG3_AC,
                O_AC_MG3 <= MAX_O_AC_MG3 * I_MG3_AC,                     

                (E_MG3_DN_buy - E_MG3_DN_sell) + (E_EES_dch_MG3 - E_EES_ch_MG3) + E_CHP_MG3 + E_PV_MG3 + E_FC_MG3 - E_EC_MG3 == EL_MG3,            # Electricity load of MG3
                (H_MG3_GN_buy - H_MG3_GN_sell) + (H_HS_dch_MG3 - H_HS_ch_MG3) + H_CHP_MG3 + H_GB1_MG3 + H_GB2_MG3 - H_AC_MG3 == HL_MG3,            # Heating load of MG3
                O_EC_MG3 + O_AC_MG3 == OL_MG3,
]


# Define the problem and solve
prob = cp.Problem(objective, constraints)
prob.solve()

# Print the optimal cost
print(f'The optimal cost is: {prob.value}')
print(f'The optimal DN cost is: {sum(DNcost.value)}')
print(f'The optimal MG1 cost is: {sum(MG1cost.value)}')
print(f'The optimal MG2 cost is: {sum(MG2cost.value)}')
print(f'The optimal MG3 cost is: {sum(MG3cost.value)}')
```

```python
import numpy as np
import cvxpy as cp
import pandas as pd
import matplotlib.pyplot as plt
import os

class MG:
    def __init__(self, lambda_dict, rho, EL_MG1, HL_MG1, OL_MG1, EL_MG2, HL_MG2, OL_MG2, EL_MG3, HL_MG3, OL_MG3):          
        self.lambda_dict = lambda_dict
        self.rho = rho
        self.EL_MG1 = EL_MG1
        self.HL_MG1 = HL_MG1
        self.EL_MG2 = EL_MG2
        self.HL_MG2 = HL_MG2
        self.EL_MG3 = EL_MG3
        self.HL_MG3 = HL_MG3
        self.OL_MG1 = OL_MG1
        self.OL_MG2 = OL_MG2
        self.OL_MG3 = OL_MG3
        self.initialize_variables()        
        self.E_MG1_DN_buy = cp.Variable(24, nonneg=True)
        self.E_MG1_DN_sell = cp.Variable(24, nonneg=True) 
        self.E_MG2_DN_buy = cp.Variable(24, nonneg=True)
        self.E_MG2_DN_sell = cp.Variable(24, nonneg=True)
        self.E_MG3_DN_buy = cp.Variable(24, nonneg=True)
        self.E_MG3_DN_sell = cp.Variable(24, nonneg=True)

        self.Cnl = 0.76
        self.L = 9.7
        self.Lamb_MG_DN_buy = 0.32
        self.Lamb_MG_DN_sell= 0.28
        self.eff_GB = 0.75
        self.eff_e = 0.45
        self.eff_rec = 0.95
        self.eff_b = 0.75
        self.Lamb_MG_GN_buy = 0.3
        self.Lamb_MG_GN_sell = 0.24
        self.MIN_H_MG_GN_Tran = 0
        self.MAX_H_MG_GN_Tran = 200

        self.MIN_E_MG_DN_Tran = 0
        self.MAX_E_MG_DN_Tran = 400

        self.MIN_EES_ch = 0
        self.MAX_EES_ch = 100
        self.MIN_EES_dch = 0
        self.MAX_EES_dch = 100
        self.MIN_E_EES = 0
        self.MAX_E_EES = 300
        self.eff_EES_ch = 0.95
        self.eff_EES_dch = 0.95

        self.MIN_E_MT_MG1 = 0
        self.MAX_E_MT_MG1 = 500
        self.MIN_E_MT_MG2 = 0
        self.MAX_E_MT_MG2 = 500
        self.MIN_E_FC_MG3 = 0
        self.MAX_E_FC_MG3 = 300        
        self.eff_MT = 0.45
        self.eff_FC = 0.45

        self.MIN_E_CHP_MG1 = 0
        self.MAX_E_CHP_MG1 = 200
        self.MIN_H_CHP_MG1 = 0
        self.MAX_H_CHP_MG1 = 200
        self.MIN_E_CHP_MG2 = 0
        self.MAX_E_CHP_MG2 = 400
        self.MIN_H_CHP_MG2 = 0
        self.MAX_H_CHP_MG2 = 400
        self.MIN_E_CHP_MG3 = 0
        self.MAX_E_CHP_MG3 = 400
        self.MIN_H_CHP_MG3 = 0
        self.MAX_H_CHP_MG3 = 400
        self.V = 0.5

        self.MIN_H_GB_MG = 0
        self.MAX_H_GB_MG = 400

        self.MIN_HS_ch = 0
        self.MAX_HS_ch = 100
        self.MIN_HS_dch = 0
        self.MAX_HS_dch = 100
        self.MIN_H_EES = 0
        self.MAX_H_EES = 300
        self.eff_HS_ch = 0.9
        self.eff_HS_dch = 0.9

        self.MIN_H_HS = 0
        self.MAX_H_HS = 300

        self.MIN_O_EC_MG1 = 0
        self.MAX_O_EC_MG1 = 400
        self.MIN_O_EC_MG2 = 0
        self.MAX_O_EC_MG2 = 400
        self.MIN_O_EC_MG3 = 0
        self.MAX_O_EC_MG3 = 400
        self.MIN_O_AC_MG1 = 0
        self.MAX_O_AC_MG1 = 400
        self.MIN_O_AC_MG2 = 0
        self.MAX_O_AC_MG2 = 400
        self.MIN_O_AC_MG3 = 0
        self.MAX_O_AC_MG3 = 400
        self.eff_EC = 0.85
        self.eff_AC = 0.85

    def initialize_variables(self):
            # Define Binary variable
            self.I_MG1_EES = cp.Variable(24, boolean=True)
            self.I_MG1_CHP = cp.Variable(24, boolean=True)
            self.I_MG1_GB1 = cp.Variable(24, boolean=True)
            self.I_MG1_GB2 = cp.Variable(24, boolean=True)
            self.I_MG1_MT = cp.Variable(24, boolean=True)
            self.I_MG1_HS = cp.Variable(24, boolean=True)
            self.I_H_MG1_GN_Tran = cp.Variable(24, boolean=True)
            self.I_E_MG1_DN_Tran = cp.Variable(24, boolean=True)
            self.I_MG1_EC = cp.Variable(24, boolean=True)
            self.I_MG1_AC = cp.Variable(24, boolean=True) 

            self.E_EES_MG1 = cp.Variable(25, nonneg=True)  # EES
            self.E_EES_ch_MG1 = cp.Variable(24, nonneg=True)
            self.E_EES_dch_MG1 = cp.Variable(24, nonneg=True)

            self.H_GB1_MG1 = cp.Variable(24, nonneg=True)  # Gas consumption cost
            self.H_GB2_MG1 = cp.Variable(24, nonneg=True)
            self.E_CHP_MG1 = cp.Variable(24, nonneg=True)
            self.H_CHP_MG1 = cp.Variable(24, nonneg=True)
            self.E_MT_MG1 = cp.Variable(24, nonneg=True)

            self.H_MG1_GN_buy = cp.Variable(24, nonneg=True)  # Heating transaction cost
            self.H_MG1_GN_sell = cp.Variable(24, nonneg=True)

            self.H_HS_MG1 = cp.Variable(25, nonneg=True)  # HS
            self.H_HS_ch_MG1 = cp.Variable(24, nonneg=True)
            self.H_HS_dch_MG1 = cp.Variable(24, nonneg=True)

            self.E_EC_MG1 = cp.Variable(24, nonneg=True) #EC
            self.O_EC_MG1 = cp.Variable(24, nonneg=True)

            self.H_AC_MG1 = cp.Variable(24, nonneg=True) #AC
            self.O_AC_MG1 = cp.Variable(24, nonneg=True) 

            # Define Binary variable
            self.I_MG2_MT = cp.Variable(24, boolean=True) 
            self.I_MG2_EES = cp.Variable(24, boolean=True)  
            self.I_MG2_CHP = cp.Variable(24, boolean=True)
            self.I_MG2_GB1 = cp.Variable(24, boolean=True)
            self.I_MG2_GB2 = cp.Variable(24, boolean=True)
            self.I_MG2_HS = cp.Variable(24, boolean=True)
            self.I_H_MG2_GN_Tran = cp.Variable(24, boolean=True)
            self.I_E_MG2_DN_Tran = cp.Variable(24, boolean=True)
            self.I_MG2_EC = cp.Variable(24, boolean=True)
            self.I_MG2_AC = cp.Variable(24, boolean=True) 

            # Define Variables                           
            self.E_MG2_DN_sell = cp.Variable(24, nonneg=True)                    # Electricity transaction cost
            self.E_MG2_DN_buy = cp.Variable(24, nonneg=True)
                            
            self.E_EES_MG2 = cp.Variable(25, nonneg=True)                        # EES
            self.E_EES_ch_MG2 = cp.Variable(24, nonneg=True)
            self.E_EES_dch_MG2 = cp.Variable(24, nonneg=True)

            # self.E_MT_MG2 = cp.Variable(24, nonneg=True)                         #  Gas consumption cost
            self.H_GB1_MG2 = cp.Variable(24, nonneg=True)
            self.H_GB2_MG2 = cp.Variable(24, nonneg=True)
            self.E_CHP_MG2 = cp.Variable(24, nonneg=True)
            self.H_CHP_MG2 = cp.Variable(24, nonneg=True)
                            
            self.H_MG2_GN_buy = cp.Variable(24, nonneg=True)                     # Heating transaction cost
            self.H_MG2_GN_sell = cp.Variable(24, nonneg=True)
                                
            self.H_HS_MG2 = cp.Variable(25, nonneg=True)                         # HS
            self.H_HS_ch_MG2 = cp.Variable(24, nonneg=True)
            self.H_HS_dch_MG2 = cp.Variable(24, nonneg=True)

            # self.H_S_MG2 = cp.Variable(24, nonneg=True)
            # self.E_S_MG2 = cp.Variable(24, nonneg=True)

            self.E_EC_MG2 = cp.Variable(24, nonneg=True) #EC
            self.O_EC_MG2 = cp.Variable(24, nonneg=True)

            self.H_AC_MG2 = cp.Variable(24, nonneg=True) #AC
            self.O_AC_MG2 = cp.Variable(24, nonneg=True) 

            # Define Binary variable
            self.I_MG3_EES = cp.Variable(24, boolean=True)  
            self.I_MG3_CHP = cp.Variable(24, boolean=True)
            self.I_MG3_GB1 = cp.Variable(24, boolean=True)
            self.I_MG3_GB2 = cp.Variable(24, boolean=True)
            self.I_MG3_FC = cp.Variable(24, boolean=True)
            self.I_MG3_HS = cp.Variable(24, boolean=True)
            self.I_H_MG3_GN_Tran = cp.Variable(24, boolean=True)
            self.I_E_MG3_DN_Tran = cp.Variable(24, boolean=True)
            self.I_MG3_EC = cp.Variable(24, boolean=True)
            self.I_MG3_AC = cp.Variable(24, boolean=True) 

            # Define Variables                  
            self.E_MG3_DN_sell = cp.Variable(24, nonneg=True)                    # Electricity transaction cost
            self.E_MG3_DN_buy = cp.Variable(24, nonneg=True)
                            
            self.E_EES_MG3 = cp.Variable(25, nonneg=True)                        # EES
            self.E_EES_ch_MG3 = cp.Variable(24, nonneg=True)
            self.E_EES_dch_MG3 = cp.Variable(24, nonneg=True)

            self.E_FC_MG3 = cp.Variable(24, nonneg=True)

            self.H_GB1_MG3 = cp.Variable(24, nonneg=True)                        #  Gas consumption cost
            self.H_GB2_MG3 = cp.Variable(24, nonneg=True)
            self.E_CHP_MG3 = cp.Variable(24, nonneg=True)
            self.H_CHP_MG3 = cp.Variable(24, nonneg=True)

            self.H_MG3_GN_buy = cp.Variable(24, nonneg=True)                     # Heating transaction cost
            self.H_MG3_GN_sell = cp.Variable(24, nonneg=True)

            self.H_HS_MG3 = cp.Variable(25, nonneg=True)                         # HS
            self.H_HS_ch_MG3 = cp.Variable(24, nonneg=True)
            self.H_HS_dch_MG3 = cp.Variable(24, nonneg=True)

            self.E_MT_MG2 = cp.Variable(24, nonneg=True)

            self.E_EC_MG3 = cp.Variable(24, nonneg=True) #EC
            self.O_EC_MG3 = cp.Variable(24, nonneg=True)

            self.H_AC_MG3 = cp.Variable(24, nonneg=True) #AC
            self.O_AC_MG3 = cp.Variable(24, nonneg=True)             

            df = pd.read_excel(r'data.xlsx', sheet_name='PV_MG')
            self.E_PV_MG1 = df['electricity'].values
            self.E_PV_MG3 = df['electricity'].values

            df = pd.read_excel(r'data.xlsx', sheet_name='WT_MG')
            self.E_WT_MG1 = df['electricity'].values
            self.E_WT_MG2 = df['electricity'].values

    def calculate_cost(self):

        C_GB1_MG1 = (self.Cnl / self.L) * (self.H_GB1_MG1 / self.eff_GB)  # Total Cost of gas use to generate electricity in MG1
        C_GB2_MG1 = (self.Cnl / self.L) * (self.H_GB2_MG1 / self.eff_GB)
        C_MT_MG1 = (self.Cnl / self.L) * (self.E_MT_MG1 / self.eff_MT)
        C_CHP_MG1 = (self.Cnl / self.L) * ((self.E_CHP_MG1 / self.eff_e + self.H_CHP_MG1 / (self.eff_rec * (1 - self.eff_e)))) / self.eff_b  # Gas cost by CHP
        C_MG1_h_g = C_GB1_MG1 + C_GB2_MG1 + C_CHP_MG1 + C_MT_MG1

        C_MG1_h_Tran = self.Lamb_MG_GN_buy * self.H_MG1_GN_buy - self.Lamb_MG_GN_sell * self.H_MG1_GN_sell  # Heat transaction cost

        C_MG1_h = C_MG1_h_g + C_MG1_h_Tran  # Heating cost for MG1

        MG1cost = C_MG1_h  # Total cost                      

        # C_MG2_e_Shortage = 1000 * self.E_S_MG2

        C_GB1_MG2 = (self.Cnl / self.L) * (self.H_GB1_MG2 / self.eff_GB)  # Total Cost of gas use to generate electricity in MG2
        C_GB2_MG2 = (self.Cnl / self.L) * (self.H_GB2_MG2 / self.eff_GB)
        C_MT_MG2 = (self.Cnl / self.L) * (self.E_MT_MG2 / self.eff_MT)
        C_CHP_MG2 = (self.Cnl / self.L) * ((self.E_CHP_MG2 / self.eff_e + self.H_CHP_MG2 / (self.eff_rec * (1 - self.eff_e)))) / self.eff_b  # Gas cost by CHP
        C_MG2_h_g = C_GB1_MG2 + C_GB2_MG2 + C_CHP_MG2 + C_MT_MG2

        C_MG2_h_Tran = self.Lamb_MG_GN_buy * self.H_MG2_GN_buy - self.Lamb_MG_GN_sell * self.H_MG2_GN_sell  # Heat transaction cost

        # C_MG2_h_Shortage = 1000 * self.H_S_MG2

        # C_MG2_e = C_MG2_e_Shortage  # Electricity cost for MG2
        C_MG2_h = C_MG2_h_g + C_MG2_h_Tran #+ C_MG2_h_Shortage  # Heating cost for MG2

        # MG2cost = C_MG2_e + C_MG2_h  # Total cost          
        MG2cost = C_MG2_h  # Total cost 

        C_GB1_MG3 = (self.Cnl / self.L) * (self.H_GB1_MG3 / self.eff_GB)  # Total Cost of gas use to generate electricity in MG3
        C_GB2_MG3 = (self.Cnl / self.L) * (self.H_GB2_MG3 / self.eff_GB)
        C_FC_MG3 = (self.Cnl / self.L) * (self.E_FC_MG3 / self.eff_FC)
        C_CHP_MG3 = (self.Cnl / self.L) * ((self.E_CHP_MG3 / self.eff_e + self.H_CHP_MG3 / (self.eff_rec * (1 - self.eff_e)))) / self.eff_b  # Gas cost by CHP
        C_MG3_h_g = C_GB1_MG3 + C_GB2_MG3 + C_CHP_MG3 + C_FC_MG3

        C_MG3_h_Tran = self.Lamb_MG_GN_buy * self.H_MG3_GN_buy - self.Lamb_MG_GN_sell * self.H_MG3_GN_sell  # Heat transaction cost

        C_MG3_h = C_MG3_h_g + C_MG3_h_Tran  # Heating cost for MG3

        MG3cost = C_MG3_h
        return MG1cost, MG2cost, MG3cost
                
    def build_constraints(self):
        constraints = [
            self.E_MG1_DN_buy >= self.MIN_E_MG_DN_Tran * self.I_E_MG1_DN_Tran,
            self.E_MG1_DN_buy <= self.MAX_E_MG_DN_Tran * self.I_E_MG1_DN_Tran,
            self.E_MG1_DN_sell >= self.MIN_E_MG_DN_Tran * (1 - self.I_E_MG1_DN_Tran),
            self.E_MG1_DN_sell <= self.MAX_E_MG_DN_Tran * (1 - self.I_E_MG1_DN_Tran),

            self.E_EES_ch_MG1 >= self.MIN_EES_ch * self.I_MG1_EES,
            self.E_EES_ch_MG1 <= self.MAX_EES_ch * self.I_MG1_EES,
            self.E_EES_dch_MG1 >= self.MIN_EES_dch * (1 - self.I_MG1_EES),
            self.E_EES_dch_MG1 <= self.MAX_EES_dch * (1 - self.I_MG1_EES),
            self.E_EES_MG1 >= self.MIN_E_EES,
            self.E_EES_MG1 <= self.MAX_E_EES,
            self.E_EES_MG1[0] == 0,
            self.E_EES_MG1[1:] == self.E_EES_MG1[:-1] + self.eff_EES_ch * self.E_EES_ch_MG1 - self.E_EES_dch_MG1 / self.eff_EES_dch,

            self.E_CHP_MG1 >= self.MIN_E_CHP_MG1 * self.I_MG1_CHP,
            self.E_CHP_MG1 <= self.MAX_E_CHP_MG1 * self.I_MG1_CHP,
            self.H_CHP_MG1 >= self.MIN_H_CHP_MG1 * self.I_MG1_CHP,
            self.H_CHP_MG1 <= self.MAX_H_CHP_MG1 * self.I_MG1_CHP,
            self.E_CHP_MG1 / self.V == self.H_CHP_MG1 / (1 - self.V),

            self.H_GB1_MG1 >= self.MIN_H_GB_MG * self.I_MG1_GB1,
            self.H_GB1_MG1 <= self.MAX_H_GB_MG * self.I_MG1_GB1,

            self.H_GB2_MG1 >= self.MIN_H_GB_MG * self.I_MG1_GB2,
            self.H_GB2_MG1 <= self.MAX_H_GB_MG * self.I_MG1_GB2,

            self.E_MT_MG1 >= self.MIN_E_MT_MG1 * self.I_MG1_MT,
            self.E_MT_MG1 <= self.MAX_E_MT_MG1 * self.I_MG1_MT,

            self.H_MG1_GN_buy >= self.MIN_H_MG_GN_Tran * self.I_H_MG1_GN_Tran,
            self.H_MG1_GN_buy <= self.MAX_H_MG_GN_Tran * self.I_H_MG1_GN_Tran,
            self.H_MG1_GN_sell >= self.MIN_H_MG_GN_Tran * (1 - self.I_H_MG1_GN_Tran),
            self.H_MG1_GN_sell <= self.MAX_H_MG_GN_Tran * (1 - self.I_H_MG1_GN_Tran),

            self.H_HS_ch_MG1 >= self.MIN_HS_ch * self.I_MG1_HS,
            self.H_HS_ch_MG1 <= self.MAX_HS_ch * self.I_MG1_HS,
            self.H_HS_dch_MG1 >= self.MIN_HS_dch * (1 - self.I_MG1_HS),
            self.H_HS_dch_MG1 <= self.MAX_HS_dch * (1 - self.I_MG1_HS),
            self.H_HS_MG1[0] == 0,
            self.H_HS_MG1[1:] == self.H_HS_MG1[:-1] + self.eff_HS_ch * self.H_HS_ch_MG1 - self.H_HS_dch_MG1 / self.eff_HS_dch,
            self.H_HS_MG1 >= self.MIN_H_HS,
            self.H_HS_MG1 <= self.MAX_H_HS, 

            self.O_EC_MG1 == self.E_EC_MG1 / self.eff_EC,
            self.O_EC_MG1 >= self.MIN_O_EC_MG1 * self.I_MG1_EC,
            self.O_EC_MG1 <= self.MAX_O_EC_MG1 * self.I_MG1_EC,

            self.O_AC_MG1 == self.H_AC_MG1 / self.eff_AC,
            self.O_AC_MG1 >= self.MIN_O_AC_MG1 * self.I_MG1_AC,
            self.O_AC_MG1 <= self.MAX_O_AC_MG1 * self.I_MG1_AC,         

            (self.E_MG1_DN_buy - self.E_MG1_DN_sell) + (self.E_EES_dch_MG1 - self.E_EES_ch_MG1) + self.E_CHP_MG1 + self.E_PV_MG1 
            + self.E_WT_MG1 + self.E_MT_MG1 - self.E_EC_MG1 == self.EL_MG1,
            (self.H_MG1_GN_buy - self.H_MG1_GN_sell) + (self.H_HS_dch_MG1 - self.H_HS_ch_MG1) + self.H_CHP_MG1 + self.H_GB1_MG1 
            + self.H_GB2_MG1 - self.H_AC_MG1 == self.HL_MG1,
            self.O_EC_MG1 + self.O_AC_MG1 == self.OL_MG1,

            self.E_MG2_DN_buy >= self.MIN_E_MG_DN_Tran * self.I_E_MG2_DN_Tran,
            self.E_MG2_DN_buy <= self.MAX_E_MG_DN_Tran * self.I_E_MG2_DN_Tran,
            self.E_MG2_DN_sell >= self.MIN_E_MG_DN_Tran * (1 - self.I_E_MG2_DN_Tran),
            self.E_MG2_DN_sell <= self.MAX_E_MG_DN_Tran * (1 - self.I_E_MG2_DN_Tran),

            self.E_EES_ch_MG2 >= self.MIN_EES_ch * self.I_MG2_EES,
            self.E_EES_ch_MG2 <= self.MAX_EES_ch * self.I_MG2_EES,
            self.E_EES_dch_MG2 >= self.MIN_EES_dch * (1 - self.I_MG2_EES),
            self.E_EES_dch_MG2 <= self.MAX_EES_dch * (1 - self.I_MG2_EES),
            self.E_EES_MG2[0] == 0,
            self.E_EES_MG2[1:] == self.E_EES_MG2[:-1] + self.eff_EES_ch * self.E_EES_ch_MG2 - self.E_EES_dch_MG2 / self.eff_EES_dch,
            self.E_EES_MG2 >= self.MIN_E_EES,
            self.E_EES_MG2 <= self.MAX_E_EES,

            self.E_MT_MG2 >= self.MIN_E_MT_MG2 * self.I_MG2_MT,
            self.E_MT_MG2 <= self.MAX_E_MT_MG2 * self.I_MG2_MT,

            self.E_CHP_MG2 >= self.MIN_E_CHP_MG2 * self.I_MG2_CHP,
            self.E_CHP_MG2 <= self.MAX_E_CHP_MG2 * self.I_MG2_CHP,
            self.H_CHP_MG2 >= self.MIN_H_CHP_MG2 * self.I_MG2_CHP,
            self.H_CHP_MG2 <= self.MAX_H_CHP_MG2 * self.I_MG2_CHP,
            self.E_CHP_MG2 / self.V == self.H_CHP_MG2 / (1 - self.V),

            self.H_GB1_MG2 >= self.MIN_H_GB_MG * self.I_MG2_GB1,
            self.H_GB1_MG2 <= self.MAX_H_GB_MG * self.I_MG2_GB1,
            self.H_GB2_MG2 >= self.MIN_H_GB_MG * self.I_MG2_GB2,
            self.H_GB2_MG2 <= self.MAX_H_GB_MG * self.I_MG2_GB2,

            self.H_MG2_GN_buy >= self.MIN_H_MG_GN_Tran * self.I_H_MG2_GN_Tran,
            self.H_MG2_GN_buy <= self.MAX_H_MG_GN_Tran * self.I_H_MG2_GN_Tran,
            self.H_MG2_GN_sell >= self.MIN_H_MG_GN_Tran * (1 - self.I_H_MG2_GN_Tran),
            self.H_MG2_GN_sell <= self.MAX_H_MG_GN_Tran * (1 - self.I_H_MG2_GN_Tran),

            self.H_HS_ch_MG2 >= self.MIN_HS_ch * self.I_MG2_HS,
            self.H_HS_ch_MG2 <= self.MAX_HS_ch * self.I_MG2_HS,
            self.H_HS_dch_MG2 >= self.MIN_HS_dch * (1 - self.I_MG2_HS),
            self.H_HS_dch_MG2 <= self.MAX_HS_dch * (1 - self.I_MG2_HS),
            self.H_HS_MG2[0] == 0,
            self.H_HS_MG2[1:] == self.H_HS_MG2[:-1] + self.eff_HS_ch * self.H_HS_ch_MG2 - self.H_HS_dch_MG2 / self.eff_HS_dch,
            self.H_HS_MG2 >= self.MIN_H_HS,
            self.H_HS_MG2 <= self.MAX_H_HS,

            self.O_EC_MG2 == self.E_EC_MG2 / self.eff_EC,
            self.O_EC_MG2 >= self.MIN_O_EC_MG2 * self.I_MG2_EC,
            self.O_EC_MG2 <= self.MAX_O_EC_MG2 * self.I_MG2_EC,

            self.O_AC_MG2 == self.H_AC_MG2 / self.eff_AC,
            self.O_AC_MG2 >= self.MIN_O_AC_MG2 * self.I_MG2_AC,
            self.O_AC_MG2 <= self.MAX_O_AC_MG2 * self.I_MG2_AC,  

            (self.E_MG2_DN_buy - self.E_MG2_DN_sell) + (self.E_EES_dch_MG2 - self.E_EES_ch_MG2) + self.E_CHP_MG2 + self.E_MT_MG2 
            + self.E_WT_MG2 - self.E_EC_MG2 == self.EL_MG2,
            (self.H_MG2_GN_buy - self.H_MG2_GN_sell) + (self.H_HS_dch_MG2 - self.H_HS_ch_MG2) + self.H_CHP_MG2 + self.H_GB1_MG2 
            + self.H_GB2_MG2 - self.H_AC_MG2 == self.HL_MG2,
            self.O_EC_MG2 + self.O_AC_MG2 == self.OL_MG2,

            self.E_MG3_DN_buy >= self.MIN_E_MG_DN_Tran * self.I_E_MG3_DN_Tran,
            self.E_MG3_DN_buy <= self.MAX_E_MG_DN_Tran * self.I_E_MG3_DN_Tran,
            self.E_MG3_DN_sell >= self.MIN_E_MG_DN_Tran * (1 - self.I_E_MG3_DN_Tran),
            self.E_MG3_DN_sell <= self.MAX_E_MG_DN_Tran * (1 - self.I_E_MG3_DN_Tran),

            self.E_EES_ch_MG3 >= self.MIN_EES_ch * self.I_MG3_EES,
            self.E_EES_ch_MG3 <= self.MAX_EES_ch * self.I_MG3_EES,
            self.E_EES_dch_MG3 >= self.MIN_EES_dch * (1 - self.I_MG3_EES),
            self.E_EES_dch_MG3 <= self.MAX_EES_dch * (1 - self.I_MG3_EES),
            self.E_EES_MG3[0] == 0,
            self.E_EES_MG3[1:] == self.E_EES_MG3[:-1] + self.eff_EES_ch * self.E_EES_ch_MG3 - self.E_EES_dch_MG3 / self.eff_EES_dch,
            self.E_EES_MG3 >= self.MIN_E_EES,
            self.E_EES_MG3 <= self.MAX_E_EES,

            self.E_FC_MG3 >= self.MIN_E_FC_MG3 * self.I_MG3_FC,
            self.E_FC_MG3 <= self.MAX_E_FC_MG3 * self.I_MG3_FC,

            self.E_CHP_MG3 >= self.MIN_E_CHP_MG3 * self.I_MG3_CHP,
            self.E_CHP_MG3 <= self.MAX_E_CHP_MG3 * self.I_MG3_CHP,
            self.H_CHP_MG3 >= self.MIN_H_CHP_MG3 * self.I_MG3_CHP,
            self.H_CHP_MG3 <= self.MAX_H_CHP_MG3 * self.I_MG3_CHP,
            self.E_CHP_MG3 / self.V == self.H_CHP_MG3 / (1 - self.V),

            self.H_GB1_MG3 >= self.MIN_H_GB_MG * self.I_MG3_GB1,
            self.H_GB1_MG3 <= self.MAX_H_GB_MG * self.I_MG3_GB1,
            self.H_GB2_MG3 >= self.MIN_H_GB_MG * self.I_MG3_GB2,
            self.H_GB2_MG3 <= self.MAX_H_GB_MG * self.I_MG3_GB2,

            self.H_MG3_GN_buy >= self.MIN_H_MG_GN_Tran * self.I_H_MG3_GN_Tran,
            self.H_MG3_GN_buy <= self.MAX_H_MG_GN_Tran * self.I_H_MG3_GN_Tran,
            self.H_MG3_GN_sell >= self.MIN_H_MG_GN_Tran * (1 - self.I_H_MG3_GN_Tran),
            self.H_MG3_GN_sell <= self.MAX_H_MG_GN_Tran * (1 - self.I_H_MG3_GN_Tran),

            self.H_HS_ch_MG3 >= self.MIN_HS_ch * self.I_MG3_HS,
            self.H_HS_ch_MG3 <= self.MAX_HS_ch * self.I_MG3_HS,
            self.H_HS_dch_MG3 >= self.MIN_HS_dch * (1 - self.I_MG3_HS),
            self.H_HS_dch_MG3 <= self.MAX_HS_dch * (1 - self.I_MG3_HS),
            self.H_HS_MG3[0] == 0,
            self.H_HS_MG3[1:] == self.H_HS_MG3[:-1] + self.eff_HS_ch * self.H_HS_ch_MG3 - self.H_HS_dch_MG3 / self.eff_HS_dch,
            self.H_HS_MG3 >= self.MIN_H_HS,
            self.H_HS_MG3 <= self.MAX_H_HS,

            self.O_EC_MG3 == self.E_EC_MG3 / self.eff_EC,
            self.O_EC_MG3 >= self.MIN_O_EC_MG3 * self.I_MG3_EC,
            self.O_EC_MG3 <= self.MAX_O_EC_MG3 * self.I_MG3_EC,

            self.O_AC_MG3 == self.H_AC_MG3 / self.eff_AC,
            self.O_AC_MG3 >= self.MIN_O_AC_MG3 * self.I_MG3_AC,
            self.O_AC_MG3 <= self.MAX_O_AC_MG3 * self.I_MG3_AC,  

            (self.E_MG3_DN_buy - self.E_MG3_DN_sell) + (self.E_EES_dch_MG3 - self.E_EES_ch_MG3) + self.E_CHP_MG3 + self.E_PV_MG3 
            + self.E_FC_MG3 - self.E_EC_MG3 == self.EL_MG3,
            (self.H_MG3_GN_buy - self.H_MG3_GN_sell) + (self.H_HS_dch_MG3 - self.H_HS_ch_MG3) + self.H_CHP_MG3 + self.H_GB1_MG3 
            + self.H_GB2_MG3 - self.H_AC_MG3 == self.HL_MG3,
            self.O_EC_MG3 + self.O_AC_MG3 == self.OL_MG3,
            ]       
        return constraints    
    
    def solve(self, dn, lambda_dict, rho):
        
        cost_mg1, cost_mg2, cost_mg3 = self.calculate_cost()
        constraints = self.build_constraints()

        lambda_MG1 = lambda_dict['MG1']
        lambda_MG2 = lambda_dict['MG2']
        lambda_MG3 = lambda_dict['MG3']
        
        MG1_objective = cp.sum(cost_mg1 + cp.multiply(lambda_MG1, (self.E_MG1_DN_buy - self.E_MG1_DN_sell)) 
                                        + (rho/2) * cp.sum_squares((dn.E_DN_MG1_buy - dn.E_DN_MG1_sell) - (self.E_MG1_DN_buy - self.E_MG1_DN_sell)))
        MG2_objective = cp.sum(cost_mg2 + cp.multiply(lambda_MG2, (self.E_MG2_DN_buy - self.E_MG2_DN_sell)) 
                                        + (rho/2) * cp.sum_squares((dn.E_DN_MG2_buy - dn.E_DN_MG2_sell) - (self.E_MG2_DN_buy - self.E_MG2_DN_sell)))
        MG3_objective = cp.sum(cost_mg3 + cp.multiply(lambda_MG3, (self.E_MG3_DN_buy - self.E_MG3_DN_sell)) 
                                        + (rho/2) * cp.sum_squares((dn.E_DN_MG3_buy - dn.E_DN_MG3_sell) - (self.E_MG3_DN_buy - self.E_MG3_DN_sell)))
        
        MGs_objective = cp.sum(MG1_objective + MG2_objective + MG3_objective)

        problem = cp.Problem(cp.Minimize(MGs_objective), constraints)
        problem.solve()
        MG1costs = sum(cost_mg1.value)
        MG2costs = sum(cost_mg2.value)
        MG3costs = sum(cost_mg3.value)
        
        if problem.status == "optimal":
            return {
                'MG1returns': MG1costs,
                'E_MG1_DN_buy': self.E_MG1_DN_buy.value,
                'E_MG1_DN_sell': self.E_MG1_DN_sell.value,
                'E_EES_dch_MG1': self.E_EES_dch_MG1.value,
                'E_EES_ch_MG1': self.E_EES_ch_MG1.value,
                'E_EES_MG1': self.E_EES_MG1.value,
                'E_CHP_MG1': self.E_CHP_MG1.value,
                'E_PV_MG1': self.E_PV_MG1,
                'E_WT_MG1': self.E_WT_MG1,
                'E_MT_MG1': self.E_MT_MG1.value,
                'EL_MG1': self.EL_MG1,
                'H_MG1_GN_buy': self.H_MG1_GN_buy.value,
                'H_MG1_GN_sell': self.H_MG1_GN_sell.value,
                'H_HS_dch_MG1': self.H_HS_dch_MG1.value,
                'H_HS_ch_MG1': self.H_HS_ch_MG1.value,
                'H_HS_MG1': self.H_HS_MG1.value,
                'H_CHP_MG1': self.H_CHP_MG1.value,
                'H_GB1_MG1': self.H_GB1_MG1.value,
                'H_GB2_MG1': self.H_GB2_MG1.value,
                'HL_MG1': self.HL_MG1,
                'O_EC_MG1':self.O_EC_MG1.value,
                'O_AC_MG1':self.O_AC_MG1.value,

                'MG2returns': MG2costs,
                'E_MG2_DN_buy': self.E_MG2_DN_buy.value,
                'E_MG2_DN_sell': self.E_MG2_DN_sell.value,
                'E_EES_dch_MG2': self.E_EES_dch_MG2.value,
                'E_EES_ch_MG2': self.E_EES_ch_MG2.value,
                'E_EES_MG2': self.E_EES_MG2.value,
                'E_CHP_MG2': self.E_CHP_MG2.value,
                'E_MT_MG2': self.E_MT_MG2.value,
                'E_WT_MG2': self.E_WT_MG2,                
                'EL_MG2': self.EL_MG2,
                'H_MG2_GN_buy': self.H_MG2_GN_buy.value,
                'H_MG2_GN_sell': self.H_MG2_GN_sell.value,
                'H_HS_dch_MG2': self.H_HS_dch_MG2.value,
                'H_HS_ch_MG2': self.H_HS_ch_MG2.value,
                'H_HS_MG2': self.H_HS_MG2.value,
                'H_CHP_MG2': self.H_CHP_MG2.value,
                'H_GB1_MG2': self.H_GB1_MG2.value,
                'H_GB2_MG2': self.H_GB2_MG2.value,
                'HL_MG2': self.HL_MG2,
                'O_EC_MG2':self.O_EC_MG2.value,
                'O_AC_MG2':self.O_AC_MG2.value,

                'MG3returns': MG3costs,
                'E_MG3_DN_buy': self.E_MG3_DN_buy.value,
                'E_MG3_DN_sell': self.E_MG3_DN_sell.value,
                'E_EES_dch_MG3': self.E_EES_dch_MG3.value,
                'E_EES_ch_MG3': self.E_EES_ch_MG3.value,
                'E_EES_MG3': self.E_EES_MG3.value,
                'E_CHP_MG3': self.E_CHP_MG3.value,
                'E_PV_MG3': self.E_PV_MG3,
                'E_FC_MG3': self.E_FC_MG3.value,
                'EL_MG3': self.EL_MG3,
                'H_MG3_GN_buy': self.H_MG3_GN_buy.value,
                'H_MG3_GN_sell': self.H_MG3_GN_sell.value,
                'H_HS_dch_MG3': self.H_HS_dch_MG3.value,
                'H_HS_ch_MG3': self.H_HS_ch_MG3.value,
                'H_HS_MG3': self.H_HS_MG3.value,
                'H_CHP_MG3': self.H_CHP_MG3.value,
                'H_GB1_MG3': self.H_GB1_MG3.value,
                'H_GB2_MG3': self.H_GB2_MG3.value,
                'HL_MG3': self.HL_MG3,
                'O_EC_MG3':self.O_EC_MG3.value,
                'O_AC_MG3':self.O_AC_MG3.value,
            }# Return the optimal energy transaction amounts
        elif problem.status == "infeasible":
            print("Problem is infeasible.")
        else:
            print("MGProblem is not solved.")
    
    def update_lambda_(self, new_lambda):
        self.lambda__dict = new_lambda
        # print("mg:",self.lambda_dict)

class DN:
    def __init__(self, lambda_dict, rho, EL_DN, HL_DN, OL_DN):  
        self.lambda_dict = lambda_dict
        self.rho = rho
        self.EL_DN = EL_DN
        self.HL_DN = HL_DN
        self.OL_DN = OL_DN
        self.initialize_variables()
        self.E_DN_MG1_sell = cp.Variable(24, nonneg=True)
        self.E_DN_MG1_buy = cp.Variable(24, nonneg=True)
        self.E_DN_MG2_sell = cp.Variable(24, nonneg=True)
        self.E_DN_MG2_buy = cp.Variable(24, nonneg=True)
        self.E_DN_MG3_sell = cp.Variable(24, nonneg=True)
        self.E_DN_MG3_buy = cp.Variable(24, nonneg=True)

        self.Lamb_DN_WS_buy =  0.32
        self.Lamb_DN_WS_sell =  0.28
        self.Lamb_DN_MG_buy = 0.28
        self.Lamb_DN_MG_sell = 0.32

        self.MAX_E_DN_WS_Tran = 650
        self.MIN_E_DN_WS_Tran = 0
        self.MAX_E_DN_MG1_Tran = 400
        self.MIN_E_DN_MG1_Tran = 0
        self.MAX_E_DN_MG2_Tran = 400
        self.MIN_E_DN_MG2_Tran = 0
        self.MAX_E_DN_MG3_Tran = 400
        self.MIN_E_DN_MG3_Tran = 0

        self.MAX_EES_ch = 100
        self.MIN_EES_ch = 0
        self.MAX_EES_dch = 100
        self.MIN_EES_dch = 0
        self.MAX_E_EES = 300
        self.MIN_E_EES = 0

        self.MAX_E_DG_DN = 500
        self.MIN_E_DG_DN = 0
        self.MAX_E_MT_DN = 400
        self.MIN_E_MT_DN = 0
        self.MAX_E_FC_DN = 300
        self.MIN_E_FC_DN = 0
        self.MAX_E_CHP1_DN = 300 
        self.MIN_E_CHP1_DN = 0
        self.MAX_H_CHP1_DN = 300
        self.MIN_H_CHP1_DN = 0
        self.MAX_E_CHP2_DN = 300
        self.MIN_E_CHP2_DN = 0
        self.MAX_H_CHP2_DN = 300
        self.MIN_H_CHP2_DN = 0
        self.MAX_H_GB_DN = 700
        self.MIN_H_GB_DN = 0

        self.MAX_H_DN_GN_Tran = 500
        self.MIN_H_DN_GN_Tran = 0

        self.MAX_HS_ch = 100
        self.MIN_HS_ch = 0
        self.MAX_HS_dch = 100
        self.MIN_HS_dch = 0
        self.MAX_H_HS = 300
        self.MIN_H_HS = 0

        self.eff_EES_ch = 0.95
        self.eff_EES_dch = 0.95
        self.eff_HS_ch = 0.9
        self.eff_HS_dch = 0.9

        self.MIN_O_EC_DN = 0
        self.MAX_O_EC_DN = 600
        self.MIN_O_AC_DN = 0
        self.MAX_O_AC_DN = 600
        self.eff_EC = 0.85
        self.eff_AC = 0.85

        self.a =  0.14
        self.b =  0.06
        self.c =  0.1
        self.Cnl = 0.76
        self.L =  9.7
        self.V = 0.5
        self.eff_GB =  0.75
        self.eff_MT =  0.45
        self.eff_FC =  0.45
        self.eff_e = 0.45
        self.eff_rec = 0.95
        self.eff_b =  0.75
        self.Lamb_DN_GN_buy =  0.3
        self.Lamb_DN_GN_sell = 0.24
        
        df = pd.read_excel(r'data.xlsx', sheet_name='PV_DN')  # PV
        self.E_PV_DN  = df['electricity'].values
        
        df = pd.read_excel(r'data.xlsx', sheet_name='WT_DN')  # Electricity demand
        self.E_WT_DN = df['electricity'].values

    def initialize_variables(self):
        # Define variables
        self.I_E_DN_WS_Tran = cp.Variable(24, boolean=True)
        self.I_H_DN_GN_Tran = cp.Variable(24, boolean=True)
        self.I_DN_EES = cp.Variable(24, boolean=True)
        self.I_DN_DG = cp.Variable(24, boolean=True)
        self.I_DN_MT = cp.Variable(24, boolean=True)
        self.I_DN_FC = cp.Variable(24, boolean=True)
        self.I_DN_CHP1 = cp.Variable(24, boolean=True)
        self.I_DN_CHP2 = cp.Variable(24, boolean=True)
        self.I_DN_GB1 = cp.Variable(24, boolean=True)
        self.I_DN_GB2 = cp.Variable(24, boolean=True)
        self.I_DN_HS = cp.Variable(24, boolean=True)
        self.I_E_DN_MG1_Tran = cp.Variable(24, boolean=True)
        self.I_E_DN_MG2_Tran = cp.Variable(24, boolean=True)
        self.I_E_DN_MG3_Tran = cp.Variable(24, boolean=True)
        self.I_DN_EC = cp.Variable(24, boolean=True)
        self.I_DN_AC = cp.Variable(24, boolean=True)

        self.E_DN_WS_buy = cp.Variable(24, nonneg=True)
        self.E_DN_WS_sell = cp.Variable(24, nonneg=True)

        self.E_EES_DN = cp.Variable(25, nonneg=True)
        self.E_EES_ch_DN = cp.Variable(24, nonneg=True)
        self.E_EES_dch_DN = cp.Variable(24, nonneg=True)

        self.E_DG_DN = cp.Variable(24, nonneg=True)
        self.E_MT_DN = cp.Variable(24, nonneg=True)
        self.E_FC_DN = cp.Variable(24, nonneg=True)
        self.H_GB1_DN = cp.Variable(24, nonneg=True)
        self.H_GB2_DN = cp.Variable(24, nonneg=True)
        self.E_CHP1_DN = cp.Variable(24, nonneg=True)
        self.H_CHP1_DN = cp.Variable(24, nonneg=True)
        self.E_CHP2_DN = cp.Variable(24, nonneg=True)
        self.H_CHP2_DN = cp.Variable(24, nonneg=True)

        self.H_DN_GN_buy = cp.Variable(24, nonneg=True)
        self.H_DN_GN_sell = cp.Variable(24, nonneg=True)

        self.H_HS_DN = cp.Variable(25, nonneg=True)
        self.H_HS_ch_DN = cp.Variable(24, nonneg=True)
        self.H_HS_dch_DN = cp.Variable(24, nonneg=True)

        self.E_EC_DN = cp.Variable(24, nonneg=True) #EC
        self.O_EC_DN = cp.Variable(24, nonneg=True)

        self.H_AC_DN = cp.Variable(24, nonneg=True) #AC
        self.O_AC_DN = cp.Variable(24, nonneg=True) 

    def calculate_cost(self): 
        # Define the total cost

        C_DN_WS_Tran = self.Lamb_DN_WS_buy * self.E_DN_WS_buy - self.Lamb_DN_WS_sell * self.E_DN_WS_sell  # Transaction cost between DN and WS

        C_DN_e_Tran = C_DN_WS_Tran   # Total electricity transaction cost

        C_DG_DN = self.a * self.E_DG_DN ** 2 + self.b * self.E_DG_DN + self.c  # Gas cost by DG
        C_MT_DN = (self.Cnl / self.L) * (self.E_MT_DN / self.eff_MT)  # Gas cost by MT
        C_FC_DN = (self.Cnl / self.L) * (self.E_FC_DN / self.eff_FC)  # Gas cost by FC
        C_GB1_DN = (self.Cnl / self.L) * (self.H_GB1_DN / self.eff_GB)  # Gas cost by GB
        C_GB2_DN = (self.Cnl / self.L) * (self.H_GB2_DN / self.eff_GB)  # Gas cost by GB
        C_CHP1_DN = ((self.Cnl / self.L) * ((self.E_CHP1_DN / self.eff_e + self.H_CHP1_DN / (self.eff_rec * (1 - self.eff_e)))) / self.eff_b)  # Gas cost by CHP
        C_CHP2_DN = ((self.Cnl / self.L) * ((self.E_CHP2_DN / self.eff_e + self.H_CHP2_DN / (self.eff_rec * (1 - self.eff_e)))) / self.eff_b)  # Gas cost by CHP
        C_DN_h_g = C_DG_DN + C_MT_DN + C_FC_DN + C_GB1_DN + C_GB2_DN + C_CHP1_DN + C_CHP2_DN  # Total Cost of gas use to generate electricity in DN

        C_DN_h_Tran = self.Lamb_DN_GN_buy * self.H_DN_GN_buy - self.Lamb_DN_GN_sell * self.H_DN_GN_sell  # Heat transaction cost

        C_DN_e = C_DN_e_Tran  # Electricity cost for distribution grid
        C_DN_h = C_DN_h_g + C_DN_h_Tran  # Heating cost for distribution grid
        DNcost = C_DN_e + C_DN_h
        # DNcost = C_DN_h
        return DNcost
    
    def build_constraints(self):
        # Define the constraints
        DN_constraints = [ 
            self.E_DN_WS_buy >= self.MIN_E_DN_WS_Tran * self.I_E_DN_WS_Tran,
            self.E_DN_WS_buy <= self.MAX_E_DN_WS_Tran * self.I_E_DN_WS_Tran,
            self.E_DN_WS_sell >= self.MIN_E_DN_WS_Tran * (1 - self.I_E_DN_WS_Tran),
            self.E_DN_WS_sell <= self.MAX_E_DN_WS_Tran * (1 - self.I_E_DN_WS_Tran),

            self.E_DN_MG1_buy >= self.MIN_E_DN_MG1_Tran * self.I_E_DN_MG1_Tran,
            self.E_DN_MG1_buy <= self.MAX_E_DN_MG1_Tran * self.I_E_DN_MG1_Tran,
            self.E_DN_MG1_sell >= self.MIN_E_DN_MG1_Tran * (1 - self.I_E_DN_MG1_Tran),
            self.E_DN_MG1_sell <= self.MAX_E_DN_MG1_Tran * (1 - self.I_E_DN_MG1_Tran),

            self.E_DN_MG2_buy >= self.MIN_E_DN_MG2_Tran * self.I_E_DN_MG2_Tran,
            self.E_DN_MG2_buy <= self.MAX_E_DN_MG2_Tran * self.I_E_DN_MG2_Tran,
            self.E_DN_MG2_sell >= self.MIN_E_DN_MG2_Tran * (1 - self.I_E_DN_MG2_Tran),
            self.E_DN_MG2_sell <= self.MAX_E_DN_MG2_Tran * (1 - self.I_E_DN_MG2_Tran),

            self.E_DN_MG3_buy >= self.MIN_E_DN_MG3_Tran * self.I_E_DN_MG3_Tran,
            self.E_DN_MG3_buy <= self.MAX_E_DN_MG3_Tran * self.I_E_DN_MG3_Tran,
            self.E_DN_MG3_sell >= self.MIN_E_DN_MG3_Tran * (1 - self.I_E_DN_MG3_Tran),
            self.E_DN_MG3_sell <= self.MAX_E_DN_MG3_Tran * (1 - self.I_E_DN_MG3_Tran),

            self.E_EES_ch_DN >= self.MIN_EES_ch * self.I_DN_EES,
            self.E_EES_ch_DN <= self.MAX_EES_ch * self.I_DN_EES,
            self.E_EES_dch_DN >= self.MIN_EES_dch * (1 - self.I_DN_EES),
            self.E_EES_dch_DN <= self.MAX_EES_dch * (1 - self.I_DN_EES),
            self.E_EES_DN[0] == 0,
            self.E_EES_DN[1:] == self.E_EES_DN[:-1] + self.eff_EES_ch * self.E_EES_ch_DN - self.E_EES_dch_DN / self.eff_EES_dch,
            self.E_EES_DN >= self.MIN_E_EES,
            self.E_EES_DN <= self.MAX_E_EES,

            self.E_DG_DN >= self.MIN_E_DG_DN * self.I_DN_DG,
            self.E_DG_DN <= self.MAX_E_DG_DN * self.I_DN_DG,

            self.E_MT_DN >= self.MIN_E_MT_DN * self.I_DN_MT,
            self.E_MT_DN <= self.MAX_E_MT_DN * self.I_DN_MT,

            self.E_FC_DN >= self.MIN_E_FC_DN * self.I_DN_FC,
            self.E_FC_DN <= self.MAX_E_FC_DN * self.I_DN_FC,

            self.E_CHP1_DN >= self.MIN_E_CHP1_DN * self.I_DN_CHP1,
            self.E_CHP1_DN <= self.MAX_E_CHP1_DN * self.I_DN_CHP1,
            self.H_CHP1_DN >= self.MIN_H_CHP1_DN * self.I_DN_CHP1,
            self.H_CHP1_DN <= self.MAX_H_CHP1_DN * self.I_DN_CHP1,
            self.E_CHP1_DN / self.V == self.H_CHP1_DN / (1 - self.V),

            self.E_CHP2_DN >= self.MIN_E_CHP2_DN * self.I_DN_CHP2,
            self.E_CHP2_DN <= self.MAX_E_CHP2_DN * self.I_DN_CHP2,
            self.H_CHP2_DN >= self.MIN_H_CHP2_DN * self.I_DN_CHP2,
            self.H_CHP2_DN <= self.MAX_H_CHP2_DN * self.I_DN_CHP2,
            self.E_CHP2_DN / self.V == self.H_CHP2_DN / (1 - self.V),

            self.H_GB1_DN >= self.MIN_H_GB_DN * self.I_DN_GB1,
            self.H_GB1_DN <= self.MAX_H_GB_DN * self.I_DN_GB1,

            self.H_GB2_DN >= self.MIN_H_GB_DN * self.I_DN_GB2,
            self.H_GB2_DN <= self.MAX_H_GB_DN * self.I_DN_GB2,

            self.H_DN_GN_buy >= self.MIN_H_DN_GN_Tran * self.I_H_DN_GN_Tran,
            self.H_DN_GN_buy <= self.MAX_H_DN_GN_Tran * self.I_H_DN_GN_Tran,
            self.H_DN_GN_sell >= self.MIN_H_DN_GN_Tran * (1 - self.I_H_DN_GN_Tran),
            self.H_DN_GN_sell <= self.MAX_H_DN_GN_Tran * (1 - self.I_H_DN_GN_Tran),

            self.H_HS_ch_DN >= self.MIN_HS_ch * self.I_DN_HS,
            self.H_HS_ch_DN <= self.MAX_HS_ch * self.I_DN_HS,
            self.H_HS_dch_DN >= self.MIN_HS_dch * (1 - self.I_DN_HS),
            self.H_HS_dch_DN <= self.MAX_HS_dch * (1 - self.I_DN_HS),
            self.H_HS_DN[0] == 0,
            self.H_HS_DN[1:] == self.H_HS_DN[:-1] + self.eff_HS_ch * self.H_HS_ch_DN - self.H_HS_dch_DN / self.eff_HS_dch,
            self.H_HS_DN >= self.MIN_H_HS,
            self.H_HS_DN <= self.MAX_H_HS, 

            self.O_EC_DN == self.E_EC_DN / self.eff_EC,
            self.O_EC_DN >= self.MIN_O_EC_DN * self.I_DN_EC,
            self.O_EC_DN <= self.MAX_O_EC_DN * self.I_DN_EC,

            self.O_AC_DN == self.H_AC_DN / self.eff_AC,
            self.O_AC_DN >= self.MIN_O_AC_DN * self.I_DN_AC,
            self.O_AC_DN <= self.MAX_O_AC_DN * self.I_DN_AC, 

            (self.E_DN_WS_buy - self.E_DN_WS_sell) + (self.E_DN_MG1_buy - self.E_DN_MG1_sell) + (self.E_DN_MG2_buy - self.E_DN_MG2_sell) + (self.E_DN_MG3_buy - self.E_DN_MG3_sell)
            + (self.E_EES_dch_DN - self.E_EES_ch_DN) + self.E_MT_DN + self.E_FC_DN + self.E_CHP1_DN + self.E_CHP2_DN + self.E_PV_DN + self.E_WT_DN - self.E_EC_DN == self.EL_DN, # Electricity load of DN

            (self.H_DN_GN_buy - self.H_DN_GN_sell) + (self.H_HS_dch_DN - self.H_HS_ch_DN) +
            self.H_CHP1_DN + self.H_CHP2_DN + self.H_GB1_DN + self.H_GB2_DN - self.H_AC_DN == self.HL_DN,  # Heating load of DN

            self.O_EC_DN + self.O_AC_DN == self.OL_DN,
        ]
        return DN_constraints
    
    def solve(self, MG, lambda_dict, rho):

        lambda_mg1 = lambda_dict['MG1']
        lambda_mg2 = lambda_dict['MG2']
        lambda_mg3 = lambda_dict['MG3']

        cost = self.calculate_cost()
        augmented_objective = cost

        augmented_objective += cp.multiply(lambda_mg1, (self.E_DN_MG1_buy - self.E_DN_MG1_sell))
        augmented_objective += cp.multiply(lambda_mg2, (self.E_DN_MG2_buy - self.E_DN_MG2_sell))
        augmented_objective += cp.multiply(lambda_mg3, (self.E_DN_MG3_buy - self.E_DN_MG3_sell))
        augmented_objective += (rho/2) * cp.sum_squares(((self.E_DN_MG1_buy - self.E_DN_MG1_sell) - (MG.E_MG1_DN_buy - MG.E_MG1_DN_sell))
                                                        + ((self.E_DN_MG2_buy - self.E_DN_MG2_sell) - (MG.E_MG2_DN_buy - MG.E_MG2_DN_sell))
                                                        + ((self.E_DN_MG3_buy - self.E_DN_MG3_sell) - (MG.E_MG3_DN_buy - MG.E_MG3_DN_sell)))

        constraints = self.build_constraints()

        problem = cp.Problem(cp.Minimize(cp.sum(augmented_objective)), constraints)
        problem.solve()
        DNcosts = sum(cost.value)
        if problem.status == "optimal":
            return {
                'DNreturns': DNcosts,
                'E_DN_WS_buy': self.E_DN_WS_buy.value, 
                'E_DN_WS_sell': self.E_DN_WS_sell.value,
                'E_DN_MG1_buy': self.E_DN_MG1_buy.value,
                'E_DN_MG1_sell': self.E_DN_MG1_sell.value,
                'E_DN_MG2_buy': self.E_DN_MG2_buy.value,
                'E_DN_MG2_sell': self.E_DN_MG2_sell.value,
                'E_DN_MG3_buy': self.E_DN_MG3_buy.value,
                'E_DN_MG3_sell': self.E_DN_MG3_sell.value,
                'E_EES_dch_DN': self.E_EES_dch_DN.value,
                'E_EES_ch_DN': self.E_EES_ch_DN.value,
                'E_EES_DN': self.E_EES_DN.value,
                'E_MT_DN': self.E_MT_DN.value,
                'E_FC_DN': self.E_FC_DN.value,
                'E_DG_DN': self.E_DG_DN.value,
                'E_CHP1_DN': self.E_CHP1_DN.value,
                'E_CHP2_DN': self.E_CHP2_DN.value,
                'E_PV_DN': self.E_PV_DN,
                'E_WT_DN': self.E_WT_DN,
                'EL_DN': self.EL_DN,
                'H_DN_GN_buy': self.H_DN_GN_buy.value,
                'H_DN_GN_sell': self.H_DN_GN_sell.value,
                'H_HS_dch_DN': self.H_HS_dch_DN.value, 
                'H_HS_ch_DN': self.H_HS_ch_DN.value,
                'H_HS_DN': self.H_HS_DN.value,
                'H_CHP1_DN': self.H_CHP1_DN.value,
                'H_CHP2_DN': self.H_CHP2_DN.value,
                'H_GB1_DN': self.H_GB1_DN.value,
                'H_GB2_DN': self.H_GB2_DN.value,
                'HL_DN': self.HL_DN,
                'O_EC_DN':self.O_EC_DN.value,
                'O_AC_DN':self.O_AC_DN.value,
            }
        elif problem.status == "infeasible":
            print("Problem is infeasible.")
        else:
            print("Problem is not solved.")
        
    def update_lambda_(self, new_lambda):
        self.lambda_dict = new_lambda
        # print("dn:",self.lambda_dict)

# Initialize variables
rho = 0.1
epsilon = 0
# number_of_time_steps = 24  # or however many time steps you have
# lambda_ = np.zeros(number_of_time_steps)

df = pd.read_excel(r'data.xlsx', sheet_name='EL_kW')  # Electricity demand
EL_DN = df['DN'].values
df = pd.read_excel(r'data.xlsx', sheet_name='EL_kW')
EL_MG1 = df['MG1'].values
df = pd.read_excel(r'data.xlsx', sheet_name='EL_kW')
EL_MG2 = df['MG2'].values
df = pd.read_excel(r'data.xlsx', sheet_name='EL_kW')
EL_MG3 = df['MG3'].values

df = pd.read_excel(r'data.xlsx', sheet_name='HL_kW')  # Heating demand
HL_DN = df['DN'].values
df = pd.read_excel(r'data.xlsx', sheet_name='HL_kW')
HL_MG1 = df['MG1'].values
df = pd.read_excel(r'data.xlsx', sheet_name='HL_kW')
HL_MG2 = df['MG2'].values
df = pd.read_excel(r'data.xlsx', sheet_name='HL_kW')
HL_MG3 = df['MG3'].values

df = pd.read_excel(r'data.xlsx', sheet_name='OL_kW')  # Cooling demand
OL_DN = df['DN'].values
df = pd.read_excel(r'data.xlsx', sheet_name='OL_kW')
OL_MG1 = df['MG1'].values
df = pd.read_excel(r'data.xlsx', sheet_name='OL_kW')
OL_MG2 = df['MG2'].values
df = pd.read_excel(r'data.xlsx', sheet_name='OL_kW')
OL_MG3 = df['MG3'].values

# Initialize MGs and DN
lambda_dict = {'MG1': 0, 'MG2': 0, 'MG3': 0}

mg = MG(lambda_dict, rho, EL_MG1, HL_MG1, OL_MG1, EL_MG2, HL_MG2, OL_MG2, EL_MG3, HL_MG3, OL_MG3)
dn = DN(lambda_dict=0, rho=rho, EL_DN=EL_DN, HL_DN=HL_DN, OL_DN=OL_DN)

costs_MG1 = []
costs_MG2 = []
costs_MG3 = []
costs_DN = []
costs_total = []

norm_values = []

while True:  # Repeat until convergence
    # Solve the problems for each MG and DN
    mg_result = mg.solve(dn, lambda_dict, rho)

    E_MG1_DN_buy_update = mg_result['E_MG1_DN_buy']
    E_MG1_DN_sell_update = mg_result['E_MG1_DN_sell']
    costs_MG1.append(mg_result['MG1returns'])
    E_CHP_MG1 = mg_result['E_CHP_MG1']
    E_PV_MG1 = mg_result['E_PV_MG1']
    E_WT_MG1 = mg_result['E_WT_MG1']
    E_MT_MG1 = mg_result['E_MT_MG1']
    E_MG1_DN_buy = mg_result['E_MG1_DN_buy']
    E_MG1_DN_sell = mg_result['E_MG1_DN_sell']
    E_EES_dch_MG1 = mg_result['E_EES_dch_MG1']
    E_EES_ch_MG1 = mg_result['E_EES_ch_MG1']
    E_EES_MG1 = mg_result['E_EES_MG1']
    H_CHP_MG1 = mg_result['H_CHP_MG1']
    H_GB1_MG1 = mg_result['H_GB1_MG1']
    H_GB2_MG1 = mg_result['H_GB2_MG1']
    H_HS_dch_MG1 = mg_result['H_HS_dch_MG1']
    H_HS_ch_MG1 = mg_result['H_HS_ch_MG1']
    H_HS_MG1 = mg_result['H_HS_MG1']
    H_MG1_GN_buy = mg_result['H_MG1_GN_buy']
    H_MG1_GN_sell = mg_result['H_MG1_GN_sell']
    O_EC_MG1 = mg_result['O_EC_MG1']
    O_AC_MG1 = mg_result['O_AC_MG1']

    E_MG2_DN_buy_update = mg_result['E_MG2_DN_buy']
    E_MG2_DN_sell_update = mg_result['E_MG2_DN_sell']
    costs_MG2.append(mg_result['MG2returns'])
    E_MG2_DN_buy = mg_result['E_MG2_DN_buy']
    E_MG2_DN_sell = mg_result['E_MG2_DN_sell']
    E_EES_dch_MG2 = mg_result['E_EES_dch_MG2']
    E_EES_ch_MG2 = mg_result['E_EES_ch_MG2']
    E_EES_MG2 = mg_result['E_EES_MG2']
    E_CHP_MG2 = mg_result['E_CHP_MG2']
    E_MT_MG2 = mg_result['E_MT_MG2']
    E_WT_MG2 = mg_result['E_WT_MG2']    
    H_MG2_GN_buy = mg_result['H_MG2_GN_buy']
    H_MG2_GN_sell = mg_result['H_MG2_GN_sell']
    H_HS_dch_MG2 = mg_result['H_HS_dch_MG2']
    H_HS_ch_MG2 = mg_result['H_HS_ch_MG2']
    H_HS_MG2 = mg_result['H_HS_MG2']
    H_CHP_MG2 = mg_result['H_CHP_MG2']
    H_GB1_MG2 = mg_result['H_GB1_MG2']
    H_GB2_MG2 = mg_result['H_GB2_MG2'] 
    # E_S_MG2 = mg_result['E_S_MG2']
    # H_S_MG2 = mg_result['H_S_MG2']
    O_EC_MG2 = mg_result['O_EC_MG2']
    O_AC_MG2 = mg_result['O_AC_MG2']  

    E_MG3_DN_buy_update = mg_result['E_MG3_DN_buy']
    E_MG3_DN_sell_update = mg_result['E_MG3_DN_sell']
    costs_MG3.append(mg_result['MG3returns'])
    E_MG3_DN_buy = mg_result['E_MG3_DN_buy']
    E_MG3_DN_sell = mg_result['E_MG3_DN_sell']
    E_EES_dch_MG3 = mg_result['E_EES_dch_MG3']
    E_EES_ch_MG3 = mg_result['E_EES_ch_MG3']
    E_EES_MG3 = mg_result['E_EES_MG3']
    E_CHP_MG3 = mg_result['E_CHP_MG3']
    E_PV_MG3 = mg_result['E_PV_MG3']
    E_FC_MG3 = mg_result['E_FC_MG3']
    H_MG3_GN_buy = mg_result['H_MG3_GN_buy']
    H_MG3_GN_sell = mg_result['H_MG3_GN_sell']
    H_HS_dch_MG3 = mg_result['H_HS_dch_MG3']
    H_HS_ch_MG3 = mg_result['H_HS_ch_MG3']
    H_HS_MG3 = mg_result['H_HS_MG3']
    H_CHP_MG3 = mg_result['H_CHP_MG3']
    H_GB1_MG3 = mg_result['H_GB1_MG3']
    H_GB2_MG3 = mg_result['H_GB2_MG3']
    O_EC_MG3 = mg_result['O_EC_MG3']
    O_AC_MG3 = mg_result['O_AC_MG3']    

    dn_result = dn.solve(mg, lambda_dict, rho)

    costs_DN.append(dn_result['DNreturns'])
    E_DN_WS_buy = dn_result['E_DN_WS_buy']
    E_DN_WS_sell = dn_result['E_DN_WS_sell']
    E_DN_MG1_buy = dn_result['E_DN_MG1_buy']
    E_DN_MG1_sell = dn_result['E_DN_MG1_sell']
    E_DN_MG2_buy = dn_result['E_DN_MG2_buy']
    E_DN_MG2_sell = dn_result['E_DN_MG2_sell']
    E_DN_MG3_buy = dn_result['E_DN_MG3_buy']
    E_DN_MG3_sell = dn_result['E_DN_MG3_sell']
    E_EES_dch_DN = dn_result['E_EES_dch_DN']
    E_EES_ch_DN = dn_result['E_EES_ch_DN']
    E_EES_DN = dn_result['E_EES_DN']
    E_MT_DN = dn_result['E_MT_DN']
    E_FC_DN = dn_result['E_FC_DN']
    E_DG_DN = dn_result['E_DG_DN']
    E_CHP1_DN = dn_result['E_CHP1_DN']
    E_CHP2_DN = dn_result['E_CHP2_DN']
    E_PV_DN = dn_result['E_PV_DN']
    E_WT_DN = dn_result['E_WT_DN']
    EL_DN = dn_result['EL_DN']
    H_DN_GN_buy = dn_result['H_DN_GN_buy']
    H_DN_GN_sell = dn_result['H_DN_GN_sell']
    H_HS_dch_DN = dn_result['H_HS_dch_DN']
    H_HS_ch_DN = dn_result['H_HS_ch_DN']
    H_HS_DN = dn_result['H_HS_DN']
    H_CHP1_DN = dn_result['H_CHP1_DN']
    H_CHP2_DN = dn_result['H_CHP2_DN']
    H_GB1_DN = dn_result['H_GB1_DN']
    H_GB2_DN = dn_result['H_GB2_DN']
    HL_DN = dn_result['HL_DN']
    O_EC_DN = dn_result['O_EC_DN']
    O_AC_DN = dn_result['O_AC_DN']

    cost_total = cp.sum(costs_MG1[-1:] + costs_MG2[-1:] + costs_MG3[-1:] + costs_DN[-1:])
    costs_total.append(cost_total)

    E_MG1_buy_value_val = E_MG1_DN_buy_update
    E_MG1_sell_value_val = E_MG1_DN_sell_update
    E_MG2_buy_value_val = E_MG2_DN_buy_update
    E_MG2_sell_value_val = E_MG2_DN_sell_update
    E_MG3_buy_value_val = E_MG3_DN_buy_update
    E_MG3_sell_value_val = E_MG3_DN_sell_update

    E_DN_MG1_buy_update = dn_result['E_DN_MG1_buy']
    E_DN_MG1_sell_update = dn_result['E_DN_MG1_sell']
    E_DN_MG2_buy_update = dn_result['E_DN_MG2_buy']
    E_DN_MG2_sell_update = dn_result['E_DN_MG2_sell']
    E_DN_MG3_buy_update = dn_result['E_DN_MG3_buy']
    E_DN_MG3_sell_update = dn_result['E_DN_MG3_sell']

    E_DN_MG1_buy_value_val = E_DN_MG1_buy_update
    E_DN_MG1_sell_value_val = E_DN_MG1_sell_update
    E_DN_MG2_buy_value_val = E_DN_MG2_buy_update
    E_DN_MG2_sell_value_val = E_DN_MG2_sell_update
    E_DN_MG3_buy_value_val = E_DN_MG3_buy_update
    E_DN_MG3_sell_value_val = E_DN_MG3_sell_update

    norm_value_MG1 = np.linalg.norm((np.array(E_MG1_buy_value_val) - np.array(E_MG1_sell_value_val)) - (np.array(E_DN_MG1_buy_value_val) - np.array(E_DN_MG1_sell_value_val)))                         
    norm_value_MG2 = np.linalg.norm((np.array(E_MG2_buy_value_val) - np.array(E_MG2_sell_value_val)) - (np.array(E_DN_MG2_buy_value_val) - np.array(E_DN_MG2_sell_value_val))) 
    norm_value_MG3 = np.linalg.norm((np.array(E_MG3_buy_value_val) - np.array(E_MG3_sell_value_val)) - (np.array(E_DN_MG3_buy_value_val) - np.array(E_DN_MG3_sell_value_val)))
    norm_value = norm_value_MG1 + norm_value_MG2 + norm_value_MG3

    print("norm_value_MG1:",norm_value_MG1)
    print("norm_value_MG2:",norm_value_MG2)
    print("norm_value_MG3:",norm_value_MG3)
    print("norm_value:",norm_value)

    norm_values.append(norm_value)

    if norm_value <= epsilon:
        break  # If converged, break the loop

    lambda_dict['MG1'] += rho * ((E_DN_MG1_buy_update - E_DN_MG1_sell_update) - (E_MG1_DN_buy_update - E_MG1_DN_sell_update))
    lambda_dict['MG2'] += rho * ((E_DN_MG2_buy_update - E_DN_MG2_sell_update) - (E_MG2_DN_buy_update - E_MG2_DN_sell_update))
    lambda_dict['MG3'] += rho * ((E_DN_MG3_buy_update - E_DN_MG3_sell_update) - (E_MG3_DN_buy_update - E_MG3_DN_sell_update))

    # print("lambda_dict:", lambda_dict)

    # Update lambda_ in each MG and DN
    mg.update_lambda_(lambda_dict)
    dn.update_lambda_(lambda_dict)

# Create the folder to store the results if it doesn't exist
folder_name = 'costs'
if not os.path.exists(folder_name):
    os.makedirs(folder_name)

# Function to write costs to a file
def write_costs_to_file(file_path, costs):
    with open(file_path, 'w') as f:
        for cost in costs:
            f.write(f"{cost}\n")

# Save the costs
write_costs_to_file(os.path.join(folder_name, 'costs_MG1.txt'), costs_MG1)
write_costs_to_file(os.path.join(folder_name, 'costs_MG2.txt'), costs_MG2)
write_costs_to_file(os.path.join(folder_name, 'costs_MG3.txt'), costs_MG3)
write_costs_to_file(os.path.join(folder_name, 'costs_DN.txt'), costs_DN)
write_costs_to_file(os.path.join(folder_name, 'costs_total.txt'), costs_total)
```
